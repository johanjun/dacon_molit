````python
# 카테고리별 시각화 함수
def viz_by_category(df, x, palette='RdPu'):
  df_new = pd.DataFrame(df.groupby(['연도', x])['사고번호'].count())

  df_new['연도'] = df_new.index.get_level_values(0)
  df_new[x] = df_new.index.get_level_values(1)


  plt.figure(figsize=(18,6))
  sns.barplot(data=df_new,
              x=x,
              y='사고번호',
              hue='연도', palette=palette)
  plt.xticks(rotation=90)
  plt.ylabel('사고 건수', fontdict={'fontsize' : 15})
  plt.xlabel(x, fontdict={'fontsize' : 15})
  plt.legend(loc='best')
  plt.show()
    
````

### 1.2. 이륜 등 관련 교통사고 



<span style="color:red"> **이 아래 그래프에서 배달앱 이용금액은 필요 없고, 배달앱 이용여부는 출처와 함께 2019년 10% 수준에서 2020년 20%로 비율이 크게 늘었다고 텍스트로 언급하면 충분** </span>


배달앱 이용관련 설문조사에 따르면 배달앱 사용에 대해서 사용한다는 응답(예)이 2018년 `7.6`%, 2019년 `11.2`%, 2020년 약 `20%`로 큰 폭으로 증가하였음을 확인할 수 있습니다. [통계청 자료](https://kosis.kr/statHtml/statHtml.do?orgId=114&tblId=DT_114054_016)) 이는 사회적 거리두기 시행 이후로 재택근무, 비대면 식사 등이 원인으로 보입니다. 


배달 건수 증가는 이륜차 이용 증가로 이어집니다. 전체적으로 2018년 이후 50cc미만부터 250cc초과까지 모든 이륜차에 대한 차량 등록 신고건수는 증가해왔고, 도로에서 늘어난 이륜차는 더 많은 사고를 발생시키고 있습니다.  ([통계청 자료](https://kosis.kr/statHtml/statHtml.do?orgId=116&tblId=DT_MLTM_2042))

배달 오토바이 등 이륜 차랑 뿐만 아니라 앞서 살펴본 것처럼 전동킥보드 같은 원동기, 개인형이동수단(PM), 자전거 등의 이용이 늘면서 사고도 늘어나는 추세입니다. 그런데 단순히 사고가 늘었다고 문제가 있다고 보기는 어렵습니다. 이 두 바퀴 수단들의 문제는 사고 피해자에 있습니다. 


````python
df_new = pd.DataFrame(df_acci[df_acci['가해운전자 차종'].isin(['이륜','원동기','자전거','개인형이동수단(PM)'])])

viz_by_category(df_new, '피해운전자 차종', 'Set1')
````

앞서 살펴본 것과 같이 2020년 전체 사고에서 보행자 사고는 2019년 대비 줄었습니다. 승용, 승합차량 사고로 인한 보행자 피해는 줄어들지만 보행자와 동선이 자주 겹치는 이륜 차량은 오히려 사고가 늘어난 것을 확인할 수 있습니다. 보행자는 이륜 차량 사고에서 두번째로 높은 사고 건수를 보입니다. 

- 보행자의 정의 : 보행자는 도로를 보행하거나 노상 작업중인 자, 노상 유희중인 자, 도로에 서 있거나 누워있는 사람, 장애자용 휠체어를 타고 있거나 밀고 가는 사람, 세발자전거나 모형자동차에 타고 있는 아이 또는 이를 밀고 가는 사람, 이륜차 원동기장치자전거 자전거를 끌고 가는 사람(단, 이륜자동차, 원동기자전거, 자전거 이외의 차량을 밀고 있는 사람은 당해 차량의 운전자로 봄)
(출처 : [도로교통공단](http://taas.koroad.or.kr/sta/acs/exs/wordArngPopup.do))

시내도로에서 가장 중요한 것이 보행자의 안전임을 감안할 때, 뭔가 대책이 필요합니다. 그럼 이 이륜 차량 사고들에 대해서 좀 더 살펴보겠습니다.

````python
#이륜차 사고만 추출한 데이터
mask = (df_acci['가해운전자 차종'].isin(['이륜','원동기','자전거','개인형이동수단(PM)']))
df_acci_2wheel = df_acci.loc[mask, :]
````

````python
viz_by_category(df_acci_2wheel, '법규위반', palette='BrBG')
````

<span style="color:red"> **피해운전자 차종이랑 법규위반이랑 엮어서 보행자사고의 원인 파악 부탁드립니다.** </span>

이륜 차량 사고의 법규 위반사례를 보면 `안전운전불이행` 법규를 위반한 사례가 많고 또 사고 건수도 늘어난 것을 확인할 수 있습니다. 뿐만 아니라 전체 교통사고 중 보행자 대상 사고에서도 `안전운전불이행`이 주요한 원인이고, 가해운전자가 개인형 이동수단인 경우가 증가했으며 사고유형은 `횡단중` 혹은 `기타` (기타는 차도보행 또는 도로에 누워있는 경우로 추정) 항목이 주요한 것을 보면 이륜이나 원동기, 자전거 이용자에 대한 안전운전 교육 등이 필요하다고 보입니다.



````python
#전체 교통사고건수 중 피해운전자가 보행자인경우
df_new = pd.DataFrame(df_acci[df_acci['피해운전자 차종'].isin(['보행자'])])

viz_by_category(df_new, '사고 월', 'hsv')
viz_by_category(df_new, '가해운전자 차종', 'hsv')
viz_by_category(df_new, '구', 'hsv')
````
